# Visualizing the Titanic Disaster

### Introduction:

This exercise is based on the titanic Disaster dataset avaiable at [Kaggle](https://www.kaggle.com/c/titanic).  
To know more about the variables check [here](https://www.kaggle.com/c/titanic/data)


### Step 1. Import the necessary libraries

```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv)

### Step 3. Assign it to a variable titanic 

```python
#reading the data
titanic=pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv')
```

### Step 4. Set PassengerId as the index 

```python
#setting the passengerId as the index
titanic=titanic.set_index('PassengerId')
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
#creating a pie chart for 
plt.pie(titanic.Sex.value_counts(),labels=titanic.Sex.value_counts().index)
plt.show()
#or
titanic.Sex.value_counts().plot(kind='pie')
#or
male=titanic[titanic.Sex=='male'].count()[3]
female=titanic[titanic.Sex=='female'].count()[3]
plt.pie([male,female],labels=['male','female'])
plt.show()
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
#creating the scatter plot of fare and age for each gender 
plt.scatter(titanic[titanic.Sex=='male'].Fare,titanic[titanic.Sex=='male'].Age,c='b')
plt.scatter(titanic[titanic.Sex=='female'].Fare,titanic[titanic.Sex=='female'].Age,c='r')
#label
plt.xlabel('fare')
plt.ylabel('age')
#le
plt.legend(['male','female'])
plt.show()
```

### Step 7. How many people survived?

```python
#checking survived 
titanic[titanic.Survived!=0].Survived.count()
```

### Step 8. Create a histogram with the Fare payed

```python
#creating the histgram for fare
plt.hist(titanic.Fare)
#label
plt.xlabel('fare')
plt.ylabel('count')
plt.show()
```
