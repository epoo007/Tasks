### Step 1. Import the necessary libraries:

```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv). 

### Step 3. Assign it to a variable called tips

```python
#read data
tips=pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv')
```

### Step 4. Delete the Unnamed 0 column

```python
#droping unnamed 0 column
tips.drop(columns='Unnamed: 0')
```

### Step 5. Plot the total_bill column histogram

```python
#making a histogram for total bill
plt.hist(tips.total_bill)
#label
plt.xlabel('bill')
plt.ylabel('count')
plt.show()
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
#creating a scatter plot for total bill and tip
plt.scatter(tips.total_bill,tips.tip)
plt.xlabel('bill')
plt.ylabel('tip')
plt.show()
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
#creating the plot
plt.figure(figsize=(10,10))
plt.scatter(tips.total_bill,tips.tip,s=tips['size']*100)
#label
plt.xlabel('bill')
plt.ylabel('tip')
plt.show()
```

### Step 8. Present the relationship between days and total_bill value

```python
#relation betwen days and total bill
tips.boxplot(by='day',column='total_bill',grid=False).figure.set_size_inches(10,10)
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
#making the scatter plot
plt.figure(figsize=(10,10))
plt.scatter(tips.tip[tips.sex=='Female'],tips.day[tips.sex=='Female'],c='pink')
plt.scatter(tips.tip[tips.sex=='Male'],tips.day[tips.sex=='Male'],c='blue')
#creating label
plt.xlabel('tip')
plt.ylabel('day')
#legend
plt.legend(tips.sex,title='sex')
plt.show()
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
tips[tips.time=='Dinner'].boxplot(by='day',column='total_bill',grid=False).figure.set_size_inches(10,10)
tips[tips.time=='Lunch'].boxplot(by='day',column='total_bill',grid=False).figure.set_size_inches(10,10)
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
fig,(ax1,ax2)=plt.subplots(1,2)
ax1.set_title('lunch')
ax1.hist(tips[tips.time=='Lunch'].tip)
ax2.set_title('Dinner')
ax2.hist(tips[tips.time=='Dinner'].tip)
plt.show()
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
#creating subplot with axis(1,2)
fig,(ax1,ax2)=plt.subplots(1,2)
#setting first subplot title as Male
ax1.set_title('Male')
#creating a scatter plot for male smokers and non smokers
ax1.scatter(tips[(tips.sex=='Male')&(tips.smoker=='No')].total_bill,tips[(tips.sex=='Male')&(tips.smoker=='No')].tip,c='r')
ax1.scatter(tips[(tips.sex=='Male')&(tips.smoker=='Yes')].total_bill,tips[(tips.sex=='Male')&(tips.smoker=='Yes')].tip,c='b')
#labels first scatter
ax1.set_xlabel('bill')
ax1.set_ylabel('tip')
#legend
ax1.legend(['non smoker','smoker'])
#creating the second subplot as female
ax2.set_title('Female')
#creating the scatter plot for female smokers ans non smokers
ax2.scatter(tips[(tips.sex=='Female')&(tips.smoker=='No')].total_bill,tips[(tips.sex=='Female')&(tips.smoker=='No')].tip,c='r')
ax2.scatter(tips[(tips.sex=='Female')&(tips.smoker=='Yes')].total_bill,tips[(tips.sex=='Female')&(tips.smoker=='Yes')].tip,c='b')
#label second scatter
ax2.set_xlabel('bill')
ax2.set_ylabel('tip')
#legend
ax2.legend(['non smoker','smoker'])
plt.show()
```
