# Apple Stock

### Introduction:

We are going to use Apple's stock price.


### Step 1. Import the necessary libraries

```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/09_Time_Series/Apple_Stock/appl_1980_2014.csv)

```python
#import data
df=pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/09_Time_Series/Apple_Stock/appl_1980_2014.csv',sep=',')
```

### Step 3. Assign it to a variable apple

```python
#assign it to apple
apple=df
```

### Step 4.  Check out the type of the columns

```python
#checking the type of each column
apple.info()
```

### Step 5. Transform the Date column as a datetime type

```python
#changing the type of the date column
apple['Date']=apple['Date'].astype('datetime64[ns]')
```

### Step 6.  Set the date as the index

```python
#changing index
apple=apple.set_index('Date')
```

### Step 7.  Is there any duplicate dates?

```python
#checking for duplicates
apple.index.duplicated().sum()
```

### Step 8.  Ops...it seems the index is from the most recent date. Make the first entry the oldest date.

```python
#reindex the dates
apple=apple.reindex(apple.index.sort_values())
```

### Step 9. Get the last business day of each month

```python
#shows the last day in the month
apple.resample('M').last()
```

### Step 10.  What is the difference in days between the first day and the oldest

```python
#show the subtraction of the last day to the first day
apple.index[-1]-apple.index[0]
```

### Step 11.  How many months in the data we have?

```python
#counts how many months we have in the data
len(apple.resample('M'))
```

### Step 12. Plot the 'Adj Close' value. Set the size of the figure to 13.5 x 9 inches

```python
#plot the adj close values
apple['Adj Close'].plot().get_figure().set_size_inches(13.5,9)
```
